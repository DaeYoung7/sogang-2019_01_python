```python
%matplotlib nbagg

import matplotlib.pylab as plt
import numpy as np

X=np.linspace(-np.pi,np.pi,256)
C=np.cos(X)
plt.plot(X,C)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$','$-\pi/2$',0,'$\pi/2$','$\pi$'])
plt.yticks([-1,0,1])
```

```python
%matplotlib nbagg

import matplotlib.pylab as plt
import numpy as np

X=np.linspace(-np.pi,np.pi,256)
C=np.cos(X)
S=np.sin(X)
plt.plot(X,C,label="cosine")
plt.plot(X,S,label="sine")
plt.legend(loc=0)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$','$-\pi/2$',0,'$\pi/2$','$\pi$'])
plt.yticks([-1,0,1])
```
