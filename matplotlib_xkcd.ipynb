```python
%matplotlib nbagg

import matplotlib.pylab as plt

import numpy as np

 

X=np.linspace(-np.pi,np.pi,256)

C=np.cos(X)

with plt.xkcd():

    plt.title("XKCD style plot!!!")

    plt.plot(X,C)

    plt.plot(2*np.pi/3,np.cos(2*np.pi/3),c="r",marker="o",ms=3)

    plt.annotate(r'$0.5 HERE!$',xy=(2*np.pi/3,np.cos(2*np.pi/3)),xycoords="data",

                 xytext=(-100,-50),textcoords="offset points", 

                 arrowprops=dict(arrowstyle="->",linewidth=1,color="k"))

plt.show()
```
