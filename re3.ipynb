```python
import re,glob
print("search 이용\n\n")
a=glob.glob("*")
p1=re.compile(r'.*[.](?!bat$).*$')
p2=re.compile(r'.*[.](?!bat$|exe$).*$')
print("bat가 아닌 경우\n")
for i in a:
    m1=p1.match(i)
    if(m1):
        print(m1.group())
print("\nbat 또는 exe가 아닌 파일\n")
for i in a:
    m2=p2.match(i)
    if(m2):
        print(m2.group())
```
