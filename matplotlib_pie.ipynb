```python
import matplotlib.pylab as plt

size=[215,130,245,210]
label=['Python','C++','Ruby','Java']
explode=(0.1,0.0,0.0,0.0)
c=['gold','yellowgreen','lightcoral','lightskyblue']
plt.pie(size,labels=label,shadow=True,autopct='%.1f%%',\
        explode=explode,colors=c,startangle=140)
plt.show()
```
